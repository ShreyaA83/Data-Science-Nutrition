```python
import pandas as pd

def get_user_input(prompt, valid_options, optional=False):
    while True:
        user_input = input(prompt).lower()
        if optional and user_input == '':  # If the input is optional and empty
            return None
        elif user_input in valid_options:
            return user_input
        else:
            print("Invalid input. Please enter a valid option.")

def filter_data(df, filters):
    filtered_data = df.copy()
    for nutrient, classification in filters.items():
        if classification is not None:
            filtered_data = filtered_data[filtered_data[nutrient] == classification]
    return filtered_data

# Function to display recommendations
def display_recommendations(data):
    print("Top Recommendations:")
    print(data[['Food code', 'Main food description','Health Score']].to_string(index=False))

# Step 1: User Input
protein_class = get_user_input("Select Protein Classification (high/medium/low, or press Enter for no filter): ", ['high', 'medium', 'low'], optional=True)
carb_class = get_user_input("Select Carbohydrate Classification (high/medium/low, or press Enter for no filter): ", ['high', 'medium', 'low'], optional=True)
fat_class = get_user_input("Select Total Fat Classification (high/medium/low, or press Enter for no filter): ", ['high', 'medium', 'low'], optional=True)
sugar_class = get_user_input("Select Sugars Classification (high/medium/low, or press Enter for no filter): ", ['high', 'medium', 'low'], optional=True)
energy_class = get_user_input("Select Energy Classification (high/medium/low, or press Enter for no filter): ", ['high', 'medium', 'low'], optional=True)

# Step 2: Health Score Calculation based on Adjusted Weights for all foods
# Use the adjusted_weights directly
standardized_df['Health Score'] = standardized_df[list(adjusted_weights.keys())].mul(list(adjusted_weights.values())).sum(axis=1)

# Normalize Health Score between 1 and 100
standardized_df['Health Score'] = ((standardized_df['Health Score'] - standardized_df['Health Score'].min()) /
                                  (standardized_df['Health Score'].max() - standardized_df['Health Score'].min())) * 100
# Step 2: Filtering Data
filters = {
    'Protein Classification': protein_class,
    'Carbohydrate Classification': carb_class,
    'Total Fat Classification': fat_class,
    'Sugars, total Classification': sugar_class,
    'Energy Classification': energy_class
}

filters = {key: value for key, value in filters.items() if value is not None}

# Filter Data
filtered_data = filter_data(standardized_df, filters)

# Step 4: Display Recommendations based on the descending order of Health Score
sorted_data = filtered_data.sort_values(by='Health Score', ascending=False)
top_recommendations = sorted_data.head(14)

# Display top recommendations
display_recommendations(top_recommendations)

# Check if there are more recommendations to show
while len(sorted_data) > len(top_recommendations):
    # Ask the user if they want to see more recommendations
    show_more = input("Do you want to see more recommendations? (yes/no): ").lower()

    if show_more == 'yes':
        # Display up to 15 additional recommendations or less if available
        additional_recommendations = sorted_data.iloc[len(top_recommendations):len(top_recommendations) + 15]
        display_recommendations(additional_recommendations)
        top_recommendations = pd.concat([top_recommendations, additional_recommendations])
    else:
        print("No additional recommendations.")
        break
```

```python
import pandas as pd
def get_user_category_preferences(category_weights):
    user_category_preferences = {}
    for category, nutrient_weights in category_weights.items():
        while True:
            try:
                rating = int(input(f"How important is {category} in your diet? (Enter a rating from -15 to 15): "))
                if -15 <= rating <= 15:
                    user_category_preferences[category] = rating
                    break
                else:
                    print("Please enter a rating within the range of -15 to 15.")
            except ValueError:
                print("Invalid input. Please enter a numeric value.")
    return user_category_preferences

def calculate_user_adjusted_weights(category_weights, user_category_preferences, adjusted_weights):
    user_adjusted_weights = {}
    for category, nutrient_weights in category_weights.items():
        category_rating = user_category_preferences.get(category, 0)
        for nutrient, weight in nutrient_weights.items():
            if nutrient in adjusted_weights:
                user_adjusted_weights[nutrient] = category_rating * (weight / 10)
            else:
                user_adjusted_weights[nutrient] = weight / 10
    return user_adjusted_weights

def calculate_health_scores(df, nutrient_weights_to_use):
    temp_df = df.copy()
    temp_df['Health Score'] = (temp_df[list(nutrient_weights_to_use.keys())] * list(nutrient_weights_to_use.values())).sum(axis=1)
    temp_df['Health Score'] = ((temp_df['Health Score'] - temp_df['Health Score'].min()) /
                               (temp_df['Health Score'].max() - temp_df['Health Score'].min())) * 100
    return temp_df.sort_values(by='Health Score', ascending=False)

def display_classification_info(selected_food, num_to_display=6):
    if not selected_food.empty:
        print("\n=== Classification Information for Selected Food ===")
        selected_food = selected_food.sort_values(by='Health Score', ascending=False)
        to_display = selected_food.head(num_to_display)
        for _, row in to_display.iterrows():
            print("\n---------------------------------------------")
            print(f"Food Code: {row['Food code']}")
            print(f"Main Food Description: {row['Main food description']}")
            print(f"WWEIA Category Description: {row['WWEIA Category description']}")
            
            print("\nNutrient Information:")
            for nutrient in standardized_df.columns[standardized_df.columns.get_loc('Protein Classification'):standardized_df.columns.get_loc('Fiber, total dietary Classification') + 1]:
                print(f"  {nutrient}: {row[nutrient]}")
                
            print(f"\nHealth Score: {row['Health Score']:.2f}\n")

        remaining_rows = selected_food.iloc[num_to_display:]

        while not remaining_rows.empty:
            try:
                show_more = input("Do you want to see more data? (yes/no): ").lower()
                if show_more == 'yes':
                    next_to_display = remaining_rows.head(4)
                    for _, row in next_to_display.iterrows():
                        print("\n---------------------------------------------")
                        print(f"Food Code: {row['Food code']}")
                        print(f"Main Food Description: {row['Main food description']}")
                        print(f"WWEIA Category Description: {row['WWEIA Category description']}")
                        
                        print("\nNutrient Information:")
                        for nutrient in standardized_df.columns[standardized_df.columns.get_loc('Protein Classification'):standardized_df.columns.get_loc('Fiber, total dietary Classification') + 1]:
                            print(f"  {nutrient}: {row[nutrient]}")
                            
                        print(f"\nHealth Score: {row['Health Score']:.2f}\n")

                    remaining_rows = remaining_rows.iloc[4:]
                elif show_more == 'no':
                    break
                else:
                    raise ValueError("Please enter 'yes' or 'no'.")
            except ValueError as e:
                print(e)

    return remaining_rows



def find_best_matches(descriptions, user_input, num_matches=80):
    user_keywords = user_input.split()
    descriptions = descriptions.copy()
    descriptions['Match Count'] = 0
    
    for keyword in user_keywords:
        descriptions['Match Count'] += descriptions.apply(lambda row: keyword.lower() in row['Main food description'].lower(), axis=1)
    
    filtered_descriptions = descriptions[descriptions['Match Count'] > 0]
    sorted_food = filtered_descriptions.sort_values(by='Match Count', ascending=False)
    top_matches = sorted_food.head(num_matches)
    
    return top_matches
def main():
    offset = 0
    
    while True:
        try:
            use_custom_preferences = input("Do you want to input custom preferences for nutrient categories? (yes/no): ").lower()
            if use_custom_preferences not in ['yes', 'no']:
                raise ValueError("Please enter 'yes' or 'no'.")
        except ValueError as e:
            print(e)
            continue

        if use_custom_preferences == 'yes':
            user_category_preferences = get_user_category_preferences(category_weights)
            use_custom_weights = input("Do you want to use custom preferences for health score calculation? (yes/no): ").lower()
            if use_custom_weights == 'yes':
                user_adjusted_weights = calculate_user_adjusted_weights(category_weights, user_category_preferences, adjusted_weights)
            else:
                user_adjusted_weights = adjusted_weights
        else:
            user_adjusted_weights = adjusted_weights

        sorted_food = calculate_health_scores(standardized_df, user_adjusted_weights)
        main_food_descriptions = input("Enter Main Food Description: ").lower()
        top_matches = find_best_matches(sorted_food.iloc[offset:], main_food_descriptions, num_matches=30)
        remaining_rows = display_classification_info(top_matches)

        user_input = input("Do you want to search for another main food description? (yes/no): ").lower()
        while user_input not in ['yes', 'no']:
            print("Please enter 'yes' or 'no'.")
            user_input = input("Do you want to search for another main food description? (yes/no): ").lower()

        if user_input != 'yes':
            break

if __name__ == "__main__":
    main()
```
