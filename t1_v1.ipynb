```python
import pandas as pd



# Function to convert values and column names
def convert_to_grams(df):
    # List of columns to convert
    mg_columns = [col for col in df.columns if '(mg)' in col]
    mcg_columns = [col for col in df.columns if '(mcg)' in col]
    
    # Convert values, handling non-numeric values
    for col in mg_columns:
        df[col] = pd.to_numeric(df[col], errors='coerce') / 1000
    
    for col in mcg_columns:
        df[col] = pd.to_numeric(df[col], errors='coerce') / 1000000

    # Convert column names
    df.rename(columns=lambda x: x.replace('(mg)', '(g)').replace('(mcg)', '(g)'), inplace=True)
    
    return df

# Apply the function to your dataset
dataset = convert_to_grams(dataset)

# Check the updated DataFrame
print(dataset.head())
print(dataset.columns)
```
